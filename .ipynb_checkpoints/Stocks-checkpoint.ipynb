```python
df['Date'] = pd.to_datetime(df['Date'])
df.set_index('Date', inplace=True)
```

# Let us analyse each company's stocks 

## Apple

```python
Apple_df = df[(df['Ticker']=='AAPL')]
Apple_df.drop(['Ticker'],axis=1)
```

```python
mpf.plot(Apple_df, type='line', style='yahoo', title='Stock Price', ylabel='Price')
```

Between february and may, Apple stocks price have increased.
bullish continual pattern can be seen from february to may, this is identified by the a peak reached by an indicator

## Google

```python
Google_df = df[(df['Ticker']=='GOOG')]
Google_df.drop(['Ticker'],axis=1)
mpf.plot(Google_df, type='line', style='yahoo', title='Stock Price', ylabel='Price')
```

Between february and may, Google stocks price have decreased.
[Cup and handle](https://www.investopedia.com/terms/c/cupandhandle.asp) continual pattern can be seen from february to may, this is identified by the c shape formed between february and march, and the [flag](https://www.investopedia.com/ask/answers/013015/what-are-main-differences-between-symmetrical-triangle-pattern-and-pennant.asp) from april to may representing the handle.

## Microsft

```python
Microsoft_df = df[(df['Ticker']=='MSFT')]
Microsoft_df.drop(['Ticker'],axis=1)
mpf.plot(Microsoft_df, type='line', style='yahoo', title='Stock Price', ylabel='Price')
```

## INSIGHT
The stock price started off decreasing frm the 7 to 28th of february, since then it increased until may. Overrall the stock price increased between the period of feb to may. 

## NETFLIX

```python
Netflix_df = df[(df['Ticker']=='NFLX')]
Netflix_df.drop(['Ticker'],axis=1)
mpf.plot(Netflix_df, type='line', style='yahoo', title='Stock Price', ylabel='Price')
```

The stock price decreased from 7 feb to march 10 and the a sudden increase between march 22 to april 1st and then decreased again until may.

```python
#df['Moving Average'] = Apple_df['Close'].rolling(window=10).mean()


df['MA5'] = df.groupby('Ticker')['Close'].rolling(window=5).mean().reset_index(0, drop=True)
df['MA20'] = df.groupby('Ticker')['Close'].rolling(window=20).mean().reset_index(0, drop=True)
df['MA30'] = df.groupby('Ticker')['Close'].rolling(window=30).mean().reset_index(0, drop=True)
```
